```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
```

```python
maindf = pd.read_csv("Data.csv")
```

```python
maindf
```

```python
maindf.describe(include='all')
```

```python
x = maindf['Gender'].value_counts()
bar_color = ['maroon', 'red']
plt.bar(x.index,x,color=bar_color)
plt.title("Which Gender takes the health insurance most?", fontsize=18)
plt.xlabel("Gender")
plt.ylabel("Frequency")
plt.show()
```
